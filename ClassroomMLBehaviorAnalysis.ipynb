```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO(r"C:\Users\91901\runs\detect\yolov8n_custom19\weights\best.pt")
#model = YOLO(r"C:\Users\91901\runs\detect\yolov8n_custom11\weights\best.pt")
# Open the video file
#video_path = "C:\\Users\\91901\\SECURUS CCTV - 2 Megapixel IP Camera with Audio Classroom Solution.mp4"

#ip_camera_address = "http://192.168.53.50:8080/video"

#cap = cv2.VideoCapture(ip_camera_address)
cap = cv2.VideoCapture("C:\\Users\\91901\Downloads\\videoplayback.mp4")
if not cap.isOpened():
    print("Error: could not open video stream")
    exit()

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()

```
